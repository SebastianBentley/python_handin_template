## 03 solution part 1

```python
import bs4
import requests

#Go to this website: https://www.komplett.dk/category/21661/gaming/gaming-pc/baerbar
r = requests.get('https://www.komplett.dk/category/21661/gaming/gaming-pc/baerbar')
r.raise_for_status()
```

```python
#Use BeautifulSoup, to select the name and price for the laptops (hint: look for class="product-price" to begin with).

soup = bs4.BeautifulSoup(r.text, 'html.parser')
names = soup.select('div[class="text-content"] > h2')
prices = soup.select('span[class="product-price-now"]')

laptops = {}

for idx, val in enumerate(names):
    laptops[names[idx].getText()] = prices[idx].getText()

print(laptops)
```

```python
#Select the data from the endpoints, using a request object.
r2 = requests.get('http://localhost:5000/api/laptops/all')
r2.raise_for_status()
soup2 = bs4.BeautifulSoup(r2.text, 'html.parser')

import re
regex_names = re.compile(r'"name":.([\w ]+)')
regex_prices = re.compile(r'"price":.([\w.]+)')
endpoint_names = regex_names.findall(str(soup2))
endpoint_price = regex_prices.findall(str(soup2))
print('Matches: {}'.format(endpoint_names))
print('Matches: {}'.format(endpoint_price))
```

```python
#Make a sorted bar chart of the laptops, where x-axis is names, and y-axis is price.
#NOTE: This only works with a running flask from EX2
import matplotlib.pyplot as plt
%matplotlib inline

name_and_price = dict(zip(regex_names.findall(str(soup2)), regex_prices.findall(str(soup2))))
    
sorted_dict = {k: v for k, v in sorted(name_and_price.items(), key=lambda item: item[1])}
print(sorted_dict)
x = list(sorted_dict.keys())
y = list(sorted_dict.values())
plt.figure(figsize=(20,10))
plt.xticks(rotation=75)
plt.title("Laptop prices", fontsize=12)
plt.xlabel("name", fontsize=16)
plt.ylabel("price", fontsize=16)

plt.bar(x,y)
```
